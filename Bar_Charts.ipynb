
# Bar Chart Visualization of Disease Data:

This code snippet creates a bar chart visualizing various disease-related data, including the total number of diseases, publications in 2022, and distributions of specific disease categories. Each disease category is represented by a different color, and the bars are grouped accordingly. The chart provides a concise overview of disease-related metrics for analysis and comparison.

```python
import matplotlib.pyplot as plt

# Data
diseases = ['ICD', 'DO', 'TGA']
total_diseases = [7, 3, 8]
publications_2022 = [10, 4, 0]
cardiovascular_diseases = [0, 0, 7]
injuries_based_diseases = [3, 0, 0]
other_diseases = [4, 3, 1]
disease_association_publications = [0, 0, 1]

# Width of the bars
bar_width = 0.15

# Set the positions of the bars on the x-axis
r1 = range(len(diseases))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]
r5 = [x + bar_width for x in r4]
r6 = [x + bar_width for x in r5]
# Create the bar chart
plt.bar(r1, total_diseases, color='b', width=bar_width, edgecolor='grey', label='Total Diseases')
plt.bar(r2, publications_2022, color='g', width=bar_width, edgecolor='grey', label='Publications in 2022')
plt.bar(r3, cardiovascular_diseases, color='r', width=bar_width, edgecolor='grey', label='Cardiovascular Diseases')
plt.bar(r4, injuries_based_diseases, color='y', width=bar_width, edgecolor='grey', label='Injuries Based Diseases')
plt.bar(r5, other_diseases, color='c', width=bar_width, edgecolor='grey', label='Other Diseases')
plt.bar(r6, disease_association_publications, color='orange', width=bar_width, edgecolor='grey', label='Diseases_Disease Association')

# Add xticks on the middle of the group bars
plt.xlabel('Diseases', fontweight='bold')
plt.xticks([r + bar_width * 2 for r in range(len(diseases))], diseases)

# Add a legend
#plt.legend()
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.savefig('bar_chart.jpg', format='jpg', bbox_inches='tight')

# Show the plot
plt.show()
```


# Comparison of Dataset Citations:

This code snippet generates a bar chart comparing the number of citations for three different datasets: Disease Ontology (DO), ICD-11 for Mortality and Morbidity Statistics, and The Guideline Advantage (TGA). The bars represent the number of citations, and each dataset is labeled on the x-axis. Additionally, a table displaying the range of years for each dataset is included below the bar chart for reference.

```python
import matplotlib.pyplot as plt

# Data
datasets = ['Disease Ontology (DO)', 'ICD-11 for Mortality and Morbidity Statistics', 'The Guideline Advantage (TGA)']
citations = [216, 379, 101]
years_range = ['2018-2022', '2018-2022', '2014-2022']

# Width of the bars
bar_width = 0.5

# Set the positions of the bars on the x-axis
x = range(len(datasets))

# Create the bar chart
plt.bar(x, citations, color='b', width=bar_width, edgecolor='grey')

# Add xticks
plt.xticks(x, datasets, rotation=45, ha='right')

# Add labels and title
plt.xlabel('Datasets')
plt.ylabel('Number of Citations')
plt.title('Comparison of the Datasets in terms of Number of Citations')

# Add a table below the bar chart
plt.table(cellText=[years_range], loc='bottom')

# Save the plot as a JPEG file
plt.savefig('Table 1.jpg', format='jpg', bbox_inches='tight')

# Show the plot
plt.show()
```
